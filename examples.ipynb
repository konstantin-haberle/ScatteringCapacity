# Separation Capacity of Scattering Networks: Examples

```python
%load_ext autoreload
%autoreload 2
import numpy as np
from scattering_networks_finite_cyclic_groups import scattering_net
from sc import SC
```

## Weyl-Heisenberg frames

### Construction of scattering network

```python
M = 30
chi_hat = np.zeros(M)
chi_hat[0] = chi_hat[1] = chi_hat[2] = chi_hat[M-2] = chi_hat[M-1] = 1
R = 5
layers = [(scattering_net.WH_frame(np.fft.ifft(chi_hat),R),scattering_net.modulus2,scattering_net.identity) for l in range(3)]
myScatteringNet_WH = scattering_net(M,layers)
```

### Separation capacity computation

```python
SC.compute_sc(myScatteringNet_WH.compute_ST, M)
```

## Wavelets

### Construction of scattering network

```python
M = 30
phi_hat = np.zeros(M)
phi_hat[0] = phi_hat[1] = phi_hat[2] = phi_hat[M-1] = 1
psi = np.random.rand(M)
layers = [(scattering_net.wavelet_frame(np.fft.ifft(phi_hat),psi),scattering_net.modulus2,scattering_net.identity) for l in range(3)]
myScatteringNet_Wav = scattering_net(M,layers)
```

### Separation capacity computation

```python
SC.compute_sc(myScatteringNet_Wav.compute_ST, M)
```

## Random frame

### Construction of scattering network

```python
M = 30
frame_rand = [np.random.rand(M)+1j*np.random.rand(M) for k in range(3)]
layers = [(frame_rand,scattering_net.modulus,scattering_net.identity) for l in range(4)]
myScatteringNet_rand = scattering_net(M,layers,struct='x')
```

### Separation capacity computation

```python
SC.compute_sc(myScatteringNet_rand.compute_ST, M)
```
