# Separation Capacity of Scattering Networks: Examples

```python
%load_ext autoreload
%autoreload 2
import numpy as np
from scattering_networks_finite_cyclic_groups import scattering_net
from sc import SC
```

```python
# Compute separation capacity averaged over multiple runs
def average_sc(func, M, n_runs=10):
    values = [SC.compute_sc(func, M) for _ in range(n_runs)]
    return np.mean(values)
```

## Weyl-Heisenberg frame

```python
M = 25
chi_hat = np.zeros(M)
chi_hat[0] = chi_hat[1] = chi_hat[2] = chi_hat[M-2] = chi_hat[M-1] = 1
R = 5
Psi_WH = scattering_net.WH_frame(np.fft.ifft(chi_hat),R) 
```

### Construction of scattering networks

```python
#mod2
layers_WH_mod2 = [(Psi_WH,scattering_net.modulus2,scattering_net.identity) for l in range(4)]
myScatteringNet_WH_mod2 = scattering_net(M,layers_WH_mod2)
#mod
layers_WH_mod = [(Psi_WH,scattering_net.modulus,scattering_net.identity) for l in range(4)]
myScatteringNet_WH_mod = scattering_net(M,layers_WH_mod)
#tanh
layers_WH_tanh = [(Psi_WH,scattering_net.tanh,scattering_net.identity) for l in range(4)]
myScatteringNet_WH_tanh = scattering_net(M,layers_WH_tanh)
#sigmoid
layers_WH_sig = [(Psi_WH,scattering_net.sigmoid,scattering_net.identity) for l in range(4)]
myScatteringNet_WH_sig = scattering_net(M,layers_WH_sig)
```

### Separation capacity computations

```python
print(average_sc(myScatteringNet_WH_mod2.compute_ST, M))
print(average_sc(myScatteringNet_WH_mod.compute_ST, M))
print(average_sc(myScatteringNet_WH_tanh.compute_ST, M))
print(average_sc(myScatteringNet_WH_sig.compute_ST, M))
```

## Wavelet frame

```python
M = 25
J = 4
phi_hat = np.zeros(M)
phi_hat[0] = phi_hat[1] = phi_hat[2] = phi_hat[M-2] = phi_hat[M-1] = 1
phi = np.fft.ifft(phi_hat)
psi_hat = np.zeros(M)
psi_hat[8:18] = 1
psi = np.fft.ifft(psi_hat)
Psi_wvt = scattering_net.wavelet_frame(phi,psi,J)
```

### Construction of scattering networks

```python
#mod2
layers_wvt_mod2 = [(Psi_wvt,scattering_net.modulus2,scattering_net.identity) for l in range(4)]
myScatteringNet_wvt_mod2 = scattering_net(M,layers_wvt_mod2)
#mod
layers_wvt_mod = [(Psi_wvt,scattering_net.modulus,scattering_net.identity) for l in range(4)]
myScatteringNet_wvt_mod = scattering_net(M,layers_wvt_mod)
#tanh
layers_wvt_tanh = [(Psi_wvt,scattering_net.tanh,scattering_net.identity) for l in range(4)]
myScatteringNet_wvt_tanh = scattering_net(M,layers_wvt_tanh)
#sigmoid
layers_wvt_sig = [(Psi_wvt,scattering_net.sigmoid,scattering_net.identity) for l in range(4)]
myScatteringNet_wvt_sig = scattering_net(M,layers_wvt_sig)
```

### Separation capacity computations

```python
print(average_sc(myScatteringNet_wvt_mod2.compute_ST, M))
print(average_sc(myScatteringNet_wvt_mod.compute_ST, M))
print(average_sc(myScatteringNet_wvt_tanh.compute_ST, M))
print(average_sc(myScatteringNet_wvt_sig.compute_ST, M))
```
